```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cl = pd.DataFrame()
df_cd = pd.DataFrame()
df_cl['angles'] = [0.0000, 22.500, 45.000, 77.500, 135.00, 157.50, 180.00]
# df_cd['angles'] = [0.0000, 22.500, 45.000, 77.500, 85.000, 90.000, 95.000, 115.00, 135.00, 157.50, 180.00]
df_cd['angles'] = [0.0000, 10.000, 22.500, 65.000, 85.000, 90.000, 95.000, 120.00, 140.00, 160.00, 180.00]

df_cl['CL'] =     [0.0000, 0.4700, 0.3000, 0.0000, 0.0000, 0.1000, 0.0000]

# df_cd['CD'] =     [0.1000, 0.2500, 0.4700, 0.9000, 1.1700, 1.2000, 1.1700, 1.0000, 0.9000, 0.8500, 0.8000]
df_cd['CD'] =     [0.1000, 0.1500, 0.2000, 1.0000, 1.1800, 1.2000, 1.1800, 1.0000, 0.8500, 0.8250, 0.8000]
```

```python
p_cd = np.polyfit(df_cd['angles'], df_cd['CD'], 7)
p_cl = np.polyfit(df_cl['angles'], df_cl['CL'], 7)
print(p_cd)
print(p_cl)

plt.plot(df_cd['angles'], df_cd['CD'], 'o')
plt.plot(df_cl['angles'], df_cl['CL'], 'o')
plt.plot(np.arange(0,180), np.polyval(p_cd, np.arange(0,180)))
plt.plot(np.arange(0,180), np.polyval(p_cl, np.arange(0,180)))
plt.show()
```
