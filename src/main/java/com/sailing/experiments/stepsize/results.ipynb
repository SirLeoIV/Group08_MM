```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_Adams_exp = pd.read_csv("results/am_results_exp.csv")
rel_error_Adams_exp = data_Adams_exp["RelativeError"]
abs_error_Adams_exp = data_Adams_exp["AbsoluteError"]

data_Euler_exp = pd.read_csv("results/euler_results_exp.csv")
rel_error_Euler_exp = data_Euler_exp["RelativeError"]
abs_error_Euler_exp = data_Euler_exp["AbsoluteError"]

data_RK4_exp = pd.read_csv("results/rk_results_exp.csv")
rel_error_RK4_exp = data_RK4_exp["RelativeError"]
abs_error_RK4_exp = data_RK4_exp["AbsoluteError"]

data_Adams_sin = pd.read_csv("results/am_results_sin.csv")
rel_error_Adams_sin = data_Adams_sin["RelativeError"]
abs_error_Adams_sin = data_Adams_sin["AbsoluteError"]

data_Euler_sin = pd.read_csv("results/euler_results_sin.csv")
rel_error_Euler_sin = data_Euler_sin["RelativeError"]
abs_error_Euler_sin = data_Euler_sin["AbsoluteError"]

data_RK4_sin = pd.read_csv("results/rk_results_sin.csv")
rel_error_RK4_sin = data_RK4_sin["RelativeError"]
abs_error_RK4_sin = data_RK4_sin["AbsoluteError"]
```

```python
# Relative Error LogLog Plot
plt.figure()
plt.plot(data_Adams_exp["h"], rel_error_Adams_exp, label="Adams")
plt.plot(data_Euler_exp["h"], rel_error_Euler_exp, label="Euler")
plt.plot(data_RK4_exp["h"], rel_error_RK4_exp, label="RK4")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("h")
plt.ylabel("Relative Error")
plt.legend()
plt.title("Relative Error vs h on e^x function")
```

```python
# Absolute Error LogLog Plot
plt.figure()
plt.plot(data_Adams_exp["h"], abs_error_Adams_exp, label="Adams")
plt.plot(data_Euler_exp["h"], abs_error_Euler_exp, label="Euler")
plt.plot(data_RK4_exp["h"], abs_error_RK4_exp, label="RK4")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("h")
plt.ylabel("Absolute Error")
plt.legend()
plt.title("Absolute Error vs h on e^x function")
```

```python
# Relative Error LogLog Plot
plt.figure()
plt.plot(data_Adams_sin["h"], rel_error_Adams_sin, label="Adams")
plt.plot(data_Euler_sin["h"], rel_error_Euler_sin, label="Euler")
plt.plot(data_RK4_sin["h"], rel_error_RK4_sin, label="RK4")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("h")
plt.ylabel("Relative Error")
plt.legend()
plt.title("Relative Error vs h on sin function")
```

```python
# Absolute Error LogLog Plot
plt.figure()
plt.plot(data_Adams_sin["h"], abs_error_Adams_sin, label="Adams")
plt.plot(data_Euler_sin["h"], abs_error_Euler_sin, label="Euler")
plt.plot(data_RK4_sin["h"], abs_error_RK4_sin, label="RK4")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("h")
plt.ylabel("Absolute Error")
plt.legend()
plt.title("Absolute Error vs h on sin function")
```
